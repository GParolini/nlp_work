```python
import pandas as pd
colnames=["Number of times pregnant", "Plasma glucose concentration", "Diastolic blood pressure", "Triceps skin fold thickness", "2-Hour serum insulin", "Body mass index", "Diabetes pedigree function", "Age", "Class variable"] 
df = pd.read_csv("pima-indians-diabetes-data.csv", names=colnames, header=None)
print(df.head(10))
```

Load the data as a numpy vector with 8 input variables (col. 1 to 8) and 1 output variable (col. 9, i.e. diabetic/not diabetic).

```python
import numpy as np
dataset = np.loadtxt('pima-indians-diabetes-data.csv', delimiter=',')
X = dataset[:,0:8] #input
Y = dataset[:,8] #output
print(X)
print(Y)
```

Define the Keras model. The model is SEQUENTIAL and there are two hidden layers and one output layer. 

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import backend, layers, models

# define the keras model
model = tf.keras.Sequential()
model.add(layers.Dense(12, input_dim=8, activation='relu'))#1st hidden layer with 12 nodes, expected data with 8 dim, activation function ReLu
model.add(layers.Dense(8, activation='relu')) #2nd hidden layer with 8 nodes, ReLu function
model.add(layers.Dense(1, activation='sigmoid')) #output layer with one node (1/0), sigmoid act. func.
```

The model is compiled using Tensorflow. To compile the model we need to specify: a) the loss function that will be used to evaluate a set of weights; b) the optimizer that is used to search through different weights for the network; c) and any optional metrics we would like to collect and report during training (in this case the accuracy). binary_crossentropy is a loss function for a binary classification problem; adam is an efficient stochastic gradient descent optimizer.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

To execute the model, we must use the fit() function. For this function, we must specify the number of EPOCHS (all the training data set is examined) and the BATCH_SIZE (number of samples considered before the weights are updated).

```python
model.fit(X, Y, epochs=150, batch_size=10)
```

The model has been trained on the whole data set. Accuracy changes every time the model is run.

```python
_, accuracy = model.evaluate(X, Y)
print('Accuracy: %.2f' % (accuracy*100))
```

You can then check how the model performs predictions (diabeti/not diabetic) on the original data. This is not a best practice, as the model real test should be done with data never previously examined by the model.

```python
predictions = model.predict_classes(X)
# summarize the first 5 cases
for i in range(5):
    print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], Y[i]))
```
